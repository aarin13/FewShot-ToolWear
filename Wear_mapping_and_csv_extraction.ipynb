```python
for i in range(1,len(wear_list)+1): 
  ds1 = pd.read_csv(f'/content/drive/MyDrive/IEEE dataset/Data collection of machining pocket/W1/{i}.csv')
  ds1.columns = ["Axial Force/N","Bending Moment of X/N.m",'Bending Moment of Y/N.m', 'Torsion of Z/N.m',  'Vibration Channel 1', 'Vibration Channel 2', 'Spindle power', 'Spindle current']
  ds1["Time stamp"] = ""
  for j in range(len(ds1)):
    ds1["Time stamp"][j] = j/300
  ds1["Wear"] = np.float64(wear_list[i-1])
  dset.reset_index(inplace=True, drop=True)
  dset = pd.concat([dset, ds1], ignore_index = True)
```
